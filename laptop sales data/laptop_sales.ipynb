```python
import re

def extract_memory_info(memory_config):
    
    pattern = r'(\d+\.?\d*)TB|(\d+)GB|([^\d\s]+(?: Storage)?)'
    
    
    valid_types = {
        'Flash Storage',
        'HDD',
        'Hybrid',
        'SSD'
    }
    
    
    matches = re.findall(pattern, memory_config)
    
    if not matches:  
        return (0, "unknown")
    
    total_size_gb = 0
    storage_types = set()

    for tb, gb, storage_type in matches:
        if tb:
            total_size_gb += float(tb) * 1000  
        elif gb:
            total_size_gb += int(gb)           

        
        if storage_type in valid_types:
            storage_types.add(storage_type.strip())

    
    if 'SSD' in storage_types and 'HDD' in storage_types:
        storage_type_combined = 'SSD + HDD'
    elif 'SSD' in storage_types and 'Hybrid' in storage_types:
        storage_type_combined = 'SSD + Hybrid'
    elif 'SSD' in storage_types and len(storage_types) == 1:
        storage_type_combined = 'SSD'
    elif 'HDD' in storage_types and len(storage_types) == 1:
        storage_type_combined = 'HDD'
    elif 'Flash Storage' in storage_types and len(storage_types) == 1:
        storage_type_combined = 'Flash Storage'
    elif 'HDD' in storage_types and len(storage_types) == 2:
        storage_type_combined = 'HDD + HDD'
    else:
        storage_type_combined = "unknown"
    
    return (int(total_size_gb), storage_type_combined)

# Apply the function to the DataFrame
laptop_sales['Memory amount (GB)'] = laptop_sales['Memory'].map(lambda x: extract_memory_info(x)[0])
laptop_sales['Memory type'] = laptop_sales['Memory'].map(lambda x: extract_memory_info(x)[1])
```

## Fit our first model using RandomForestRegressor on Company, TypeName, RAM(GB), OpSys, Memory amount(GB), Memory Type

```python
X = laptop_sales.drop(['ScreenResolution', 'Inches', 'Cpu', 'Gpu', 'Memory', 'Weight (Kg)', 'Price'], axis=1)
y = laptop_sales['Price']
X
```

```python
X.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
onehot = OneHotEncoder()
cat_feat = ['OpSys','TypeName', 'Company', 'Memory type']
col_trans = ColumnTransformer([
    ('huha', onehot, cat_feat)],
    remainder='passthrough')
transformed_x = col_trans.fit_transform(X)
pd.DataFrame(transformed_x)
```

```python
np.random.seed(2)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(transformed_x,
                                                   y,
                                                   test_size=0.2)
```

```python
np.random.seed(2)
model = RandomForestRegressor()
model.fit(X_train, y_train)
model.score(X_train, y_train), model.score(X_test, y_test)
```

```python
np.random.seed(2)

from sklearn.model_selection import cross_val_score
first_csv = cross_val_score(model,
                           transformed_x,
                           y,
                           cv=5,
                           scoring=None)# this parameter becomes the estimators default score
print(first_csv)
first_csv.mean()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_model(y_true, y_preds, model_name):
    accu = r2_score(y_true, y_preds)
    mae = mean_absolute_error(y_true, y_preds)
    mse = mean_squared_error(y_true, y_preds)
    metrics_dict = {'accuracy': accu,
                   'mean absolute error' : mae,
                   'mean squared error' : mse}
    print(f'{model_name}\'s accuracy: {accu*100:.2f}')
    print(f'{model_name}\'s mean absolute error: {mae:.2f}')
    print(f'{model_name}\'s mean squared error: {mse:.2f}')
    return metrics_dict
```

```python
np.random.seed(2)

y_preds = model.predict(X_test)
evaluate_model(y_test, y_preds, 'first model')
```
